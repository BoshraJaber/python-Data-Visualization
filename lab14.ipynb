```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from calendar import month_abbr
```

```python
sns.get_dataset_names() # to load databases that exist is seaborn
penguins = sns.load_dataset("penguins")
penguins
```

```python
penguins_by_species = penguins.groupby("species", as_index=False)

sns.barplot(x="species", y="body_mass_g", data=penguins)
penguins_by_species
```

```python
sns.countplot(data=penguins, x="island")
```

```python
sns.kdeplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```
